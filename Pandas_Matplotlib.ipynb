```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("pandas_matplotlib.xlsx")
```

```python
#inspect data
# Display first few rows
print("Output-1")
print(df.head())
# Display info about data types and missing values
print("Output-2")
print(df.info())
# Column names
print("Output-3")
print(df.columns)
# Shape of dataset
print("Output-4")
print(df.shape)
```

```python
#Check for missing values
print(df.isnull().sum())
```

```python
# Fill or drop missing values
df = df.fillna(0)  # or df.dropna()
```

```python
#Rename Column
df.rename(columns={'pr3': 'p3'}, inplace=True)
df.rename(columns={'pr2': 'p2'}, inplace=True)
df.rename(columns={'pr1': 'p1'}, inplace=True)
print(df.head(1))
```

```python
# Create a new column
df['average'] = (df['p1'] + df['p2'] + df['p3']) / 3
print(df)
```

```python
# Sort data by total
df = df.sort_values(by='Total', ascending=False)
print(df)
```

```python
#round at 2 decimal places
df['average'] = df['average'].round(2)
print(df)
```

```python
# Average total score by gender
print(df.groupby('gender')['Total'].mean())
```

```python
# Average score by area
print(df.groupby('area')[['p1', 'p2', 'p3', 'Total']].mean())
```

```python
# Count of students by area
print(df['area'].value_counts())
```

```python
#Filtering with Multiple Conditions
# Students from "Urban" area with total > 250
urban_top = df[(df['area'] == 'Urban') & (df['total'] > 250)]
print(urban_top)
```

```python
#Using query() for readable filtering
df.query('gender == "Female" and total > 200')
```

```python
#Apply custom functions
# Define a grade calculator
def grade(x):
    if x >= 400: return 'A'
    elif x >= 220: return 'B'
    elif x >= 150: return 'C'
    else: return 'D'

df['grade'] = df['average'].apply(grade)
print(df)
```

```python
#Pivot Table (Excel-style summary)
pivot = pd.pivot_table(
    df,
    values='total',
    index='area',
    columns='gender',
    aggfunc='sum'
)
print(pivot)
```

```python
#Multi-level Grouping
# Mean total by both area and gender
grouped = df.groupby(['area', 'gender'])['total'].agg(['mean', 'count', 'max'])
print(grouped)
```

```python
#Sorting with multiple columns
df.sort_values(by=['area', 'total'], ascending=[True, False], inplace=True)
print(df)
```

Now we use matplotlib

```python
df.rename(columns={'Total': 'total'}, inplace=True)
```

```python
# Bar chart: Average total by gender
df.groupby('gender')['total'].mean().plot(kind='bar', color=['skyblue', 'pink'])
plt.title("Average Total by Gender")
plt.ylabel("Average Total")
plt.show()
```

```python
# Histogram: Distribution of total scores
plt.hist(df['total'], color='red', edgecolor='black' , bins=10)
plt.title("Distribution of Total Scores")
plt.xlabel("Total")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pie chart: Students by area
df['area'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Students by Area")
plt.ylabel('')
plt.show()
```

```python
# Pie chart: Students by gender
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Students by Area")
plt.ylabel('')
plt.show()
```

```python
# Scatter plot: pr1 vs total
plt.scatter(df['p1'], df['total'], color='purple')
plt.title("P1 vs Total")
plt.xlabel("P1")
plt.ylabel("Total")
plt.show()
```

```python
#Boxplot (for spread of scores)
plt.figure(figsize=(7,5))
df[['p1', 'p2', 'p3', 'total']].plot(kind='box')
plt.title("Score Distribution Across Tests")
plt.show()
```

```python
#Subplots (multiple charts together)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
df['total'].plot(kind='hist', bins=10, color='orange', ax=axes[0], title='Total Distribution')
df['area'].value_counts().plot(kind='bar', color='teal', ax=axes[1], title='Count by Area')
plt.tight_layout()
plt.show()
```

```python
#3 Line Plot for Trends
# If you had test scores as time series per student:
df[['p1', 'p2', 'p3']].mean().plot(kind='line', marker='o', color='red')
plt.title("Average Scores Across Tests")
plt.xlabel("Test")
plt.ylabel("Average Score")
plt.grid(True)
plt.show()
```

```python
# Grouped Bar Chart
avg_by_gender_area = df.groupby(['area', 'gender'])['total'].mean().unstack()
avg_by_gender_area.plot(kind='bar', figsize=(8,5))
plt.title("Average Total by Area and Gender")
plt.ylabel("Average Total")
plt.xlabel("Area")
plt.xticks(rotation=0)
plt.legend(title="Gender")
plt.show()
```
