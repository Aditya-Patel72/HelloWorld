```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel("Finance_Sheet.xlsx")
df
```

```python
#Plot Setup (Good to Add at the Start)
plt.style.use('ggplot')     # Optional: Better styling
plt.figure(figsize=(10, 6)) # Set default plot size
```

```python
df.groupby('BANK')['AMOUNT'].sum().plot(kind='bar')
plt.title('Total Loan Amount by Bank')
plt.xlabel('Bank')
plt.ylabel('Loan Amount')
plt.show()
```

```python
df['GENDER'].value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.title('Gender Distribution')
plt.show()
```

```python
#Line Chart  EMI over TIME
df.sort_values('TIME').plot(x='TIME', y='EMI', kind='line')
plt.title('EMI vs Time')
plt.xlabel('Time (Years)')
plt.ylabel('EMI')
plt.show()
```

```python
#Histogram  Loan Amount Distribution
df['AMOUNT'].plot(kind='hist', bins=10)
plt.title('Loan Amount Distribution')
plt.xlabel('Amount')
plt.show()
```

```python
#Scatter Plot  Rate vs Amount
df.plot(kind='scatter', x='Rate', y='AMOUNT')
plt.title('Rate vs Loan Amount')
plt.show()
```

```python
#Box Plot  EMI by Gender
df.boxplot(column='EMI', by='GENDER')
plt.title('EMI by Gender')
plt.suptitle('')  # Optional: remove default title
plt.show()
```

```python
#Bar Plot  Total Loan Amount by Bank
bank_loan = df.groupby('BANK')['AMOUNT'].sum()
bank_loan.plot(kind='bar', color='skyblue')
plt.title('Total Loan Amount by Bank')
plt.xlabel('Bank')
plt.ylabel('Loan Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Pie Chart  Gender Distribution
gender_counts = df['GENDER'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['pink', 'lightblue'])
plt.title('Gender Distribution')
plt.ylabel('')  # Hide y-label
plt.show()
```

```python
#Line Plot  EMI vs Time
df_sorted = df.sort_values('TIME')  # Optional: Sort by TIME
plt.plot(df_sorted['TIME'], df_sorted['EMI'], marker='o')
plt.title('EMI vs Time')
plt.xlabel('Time (Years)')
plt.ylabel('EMI')
plt.grid(True)
plt.show()
```

```python
#Scatter Plot  Rate vs Amount
plt.scatter(df['Rate'], df['AMOUNT'], alpha=0.7, color='green')
plt.title('Rate vs Loan Amount')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Loan Amount')
plt.show()
```

```python
#Histogram  Distribution of Loan Amounts
plt.hist(df['AMOUNT'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#Grouped Bar Plot  Average EMI by Bank and Gender
avg_emi = df.groupby(['BANK', 'GENDER'])['EMI'].mean().unstack()
avg_emi.plot(kind='bar')
plt.title('Average EMI by Bank and Gender')
plt.xlabel('Bank')
plt.ylabel('Average EMI')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Box Plot  EMI Distribution by Gender
df.boxplot(column='EMI', by='GENDER')
plt.title('EMI Distribution by Gender')
plt.suptitle('')  # Remove the default title
plt.xlabel('Gender')
plt.ylabel('EMI')
plt.show()
```

```python
#Subplots  Multiple Graphs Together
fig, axs = plt.subplots(1, 2, figsize=(12, 5))
# Pie Chart
df['GENDER'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axs[0], title='Gender Distribution')
# Bar Plot
df.groupby('BANK')['AMOUNT'].sum().plot(kind='bar', ax=axs[1], title='Loan Amount by Bank', color='teal')
plt.tight_layout()
plt.show()
```

```python
#Customizing Plots (Optional Styling)
plt.title('Chart Title', fontsize=14, fontweight='bold')
plt.xlabel('X-axis Label', fontsize=12)
plt.ylabel('Y-axis Label', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
```
